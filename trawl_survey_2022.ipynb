```python
# import libraries (DO NOT REMOVE)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in the csv file
df = pd.read_csv('22560_UNION_FSCS_SVCAT.csv')
stratum = pd.read_csv('stratum.csv')
```

```python
# remove last two columns
df = df.iloc[:, :-2]
# remove TOW, ID, SVSPP columns
df = df.drop(['TOW', 'ID', 'SVSPP'], axis=1)
display(df)

# only get the columns 'strgrp_desc', 'stratum', 'stratum_name'
stratum = stratum[['strgrp_desc', 'stratum', 'stratum_name']]
# change stratum column to STRATUM (for join)
stratum = stratum.rename(columns={'stratum': 'STRATUM'})
display(stratum)

# join df and stratum on STRATUM column
df = df.merge(stratum, on='STRATUM')
display(df)
```

```python
# group by LOGGED_SPECIES_NAME
silver_hake = df[df['LOGGED_SPECIES_NAME'] == 'SILVER HAKE']
butterfish = df[df['LOGGED_SPECIES_NAME'] == 'BUTTERFISH']
longfin_squid = df[df['LOGGED_SPECIES_NAME'] == 'LONGFIN SQUID']

display(silver_hake)
display(butterfish)
display(longfin_squid)
```

```python
# group df by LOGGED_SPECIES_NAME and count the number of rows, then sort by count
test = df.groupby('LOGGED_SPECIES_NAME').size().reset_index(name='counts')
test.sort_values(by=['counts'], inplace=True, ascending=False)
test
```
